# VGG19

```python
import os
import pickle
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input
from tensorflow.keras.layers import GlobalMaxPooling2D
from numpy.linalg import norm
from tqdm import tqdm
```

```python
input_shape = (224, 224, 3)

# Loading VGG19 model
base_model = VGG19(weights="imagenet", include_top=False, input_shape=input_shape)
base_model.trainable = False

# Now Add pooling layer to the model
model = tf.keras.Sequential([
    base_model,
    GlobalMaxPooling2D()
])
```

```python
# Define a function to extract features in batches
def extract_features_batch(image_paths, model):
    batch_images = []
    for path in image_paths:
        img = image.load_img(path, target_size=(input_shape[0], input_shape[1]))
        img_array = image.img_to_array(img)
        batch_images.append(img_array)
    batch_images = preprocess_input(np.array(batch_images))
    batch_features = model.predict(batch_images)
    normalized_batch_features = batch_features / norm(batch_features, axis=1, keepdims=True)
    return normalized_batch_features
```

```python
# Define image folder and batch size
img_folder = 'images _fashion_furniture'
batch_size = 32  
img_filenames = [os.path.join(img_folder, file) for file in os.listdir(img_folder)]


# Extract features in batches and store them
feature_list = []
for i in tqdm(range(0, len(img_filenames), batch_size)):
    batch_paths = img_filenames[i:i + batch_size]
    features = extract_features_batch(batch_paths, model)
    feature_list.extend(features)

# Save extracted features and image filenames
pickle.dump(feature_list, open('features_list_vgg19.pkl', 'wb'))
pickle.dump(img_filenames, open('image_filenames_vgg19.pkl', 'wb'))
```
